#  <font color='red'><center> Pr√≥bkowanie ziarnisto≈õci </center></font> 
- upsampling 
- downsampling

```python
ZbiorDzienny=wykres20.A1.resample('D').mean()
ZbiorDzienny
```

```python
ZbiorDzienny.isnull().sum()
```

```python
ZbiorDzienny=wykres20.A1.resample('D').mean()
Pelny=ZbiorDzienny.interpolate(method='linear')
plt.figure(figsize=(16, 8))
Pelny.plot()
```

```python
ZbiorDzienny=wykres20.A1.resample('D').mean()
Pelny=ZbiorDzienny.interpolate(method='spline',order=2)
plt.figure(figsize=(16, 8))
Pelny.plot()
```

```python
ZbiorDzienny=wykres20.A1.resample('Q').mean()
Pelny=ZbiorDzienny.interpolate(method='spline',order=2)
plt.figure(figsize=(16, 8))
Pelny.plot()
```

```python
ZbiorDzienny=dane.A1.resample('A').mean()
Pelny=ZbiorDzienny.interpolate(method='spline',order=2)
plt.figure(figsize=(16, 8))
Pelny.plot()
```
